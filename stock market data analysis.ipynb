```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
company_list=['AAPL_data.csv','GOOGL_data.csv']
df=pd.DataFrame()
```

```python
for i in company_list:
    df_current=pd.read_csv(i)
    df=pd.concat([df,df_current])
```

```python
df.shape
```

```python
df.head()
```

```python
plt.figure(figsize=(20,12))
name_company=df['Name'].unique()
for k,j in enumerate(name_company,1): #company name 
    plt.subplot(2,2,k)
    all_data=df[df['Name']==j]
    plt.plot(all_data['date'],all_data['close'])
    plt.title(j)
```

```python
plt.figure(figsize=(20,12))
name_company=df['Name'].unique()
for k,j in enumerate(name_company,1): #company name 
    plt.subplot(2,2,k)
    all_data=df[df['Name']==j]
    plt.plot(all_data['date'],all_data['volume'])
    plt.title(j)
```

```python
for company in (name_company):
    df1=df[df['Name']==company]
    
    fig = px.line(df1, x="date", y="volume", title=company)
    fig.show()
```

##### Daily Stock Return Formula
    To calculate how much you gained or lost per day for a stock, subtract the opening price from the closing price. Then, multiply the result by the number of shares you own in the company. 

```python
df=pd.read_csv('AAPL_data.csv')
```

```python
df.head()
```

```python
df['1day % return']=((df['close']-df['open'])/df['close'])*100
df.head()
```

```python
import plotly.express as px
fig = px.line(df, x="date", y="1day % return", title='')
fig.show()
```

```python
plt.figure(figsize=(10,6))
df['1day % return'].plot()
```

```python
df.set_index('date')['2016-01-01':'2016-03-31']['1day % return'].plot()
plt.xticks(rotation='vertical')
```
